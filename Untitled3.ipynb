# Evaluate forecast


````python
model = ARIMA(train, order=(5,1,0))
model_fit = model.fit()

forecast = model_fit.forecast(steps=len(test))
forecast = np.array(forecast)  # Convert forecast to a numpy array
forecast = forecast.reshape(-1, 1)  # Reshape forecast into a 2D array

mse = mean_squared_error(test.values.reshape(-1), forecast)
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
````

# Plot actual vs forecast


````python
plt.figure(figsize=(10, 6))
plt.plot(test.index, test, label='Actual')
plt.plot(test.index, forecast, color='red', label='Forecast')
plt.title("Actual vs Forecasted Daily Female Births in California")
plt.xlabel("Date")
plt.ylabel("Number of Births")
plt.legend()
plt.show()
````
